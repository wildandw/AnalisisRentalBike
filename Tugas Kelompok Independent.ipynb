### Pertanyaan 2: Apakah terdapat perbedaan penggunaan sepeda pada hari kerja dan hari libur ?
### 10122374 - Wildan Dwi Wijaksana


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

day_data = pd.read_csv("bike_day.csv")
hour_data = pd.read_csv("bike_hour.csv")
perbedaan_penggunaan_harilibur_dan_kerja = day_data.groupby('workingday')['cnt'].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=perbedaan_penggunaan_harilibur_dan_kerja.index, y=perbedaan_penggunaan_harilibur_dan_kerja.values)
plt.title("Perbedaan Penggunaan Sepeda pada hari kerja dan hari libur")
plt.xlabel("Hari")
plt.ylabel("Rata-Rata Jumlah Sepeda")
plt.xticks(ticks=[0, 1], labels=["Hari Libur", "Hari Kerja"])
plt.show()
```

### Pertanyaan 5 : Bagaimana Tampilan visualisasi Perbandingan Jumlah Registrasi antara Tahun Pertama dan Tahun Kedua?
### 10122351 - Rifqi Khoerul Assidiq

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV (ganti "nama_file.csv" dengan nama file yang sesuai)
data = pd.read_csv("bike_day.csv")

# Mengonversi kolom 'dteday' ke dalam format datetime
data['dteday'] = pd.to_datetime(data['dteday'])

# Mendapatkan data untuk tahun pertama dan tahun kedua
data_tahun_pertama = data[data['dteday'].dt.year == data['dteday'].dt.year.min()]
data_tahun_kedua = data[data['dteday'].dt.year == data['dteday'].dt.year.max()]

# Menghitung jumlah registrasi untuk setiap tahun
jumlah_registrasi_tahun_pertama = data_tahun_pertama['registered'].sum()
jumlah_registrasi_tahun_kedua = data_tahun_kedua['registered'].sum()

# Membuat grafik untuk membandingkan jumlah registrasi antara tahun pertama dan tahun kedua
tahun = ['Tahun Pertama', 'Tahun Kedua']
jumlah_registrasi = [jumlah_registrasi_tahun_pertama, jumlah_registrasi_tahun_kedua]

plt.figure(figsize=(8, 6))
plt.bar(tahun, jumlah_registrasi, color=['lightgreen', 'lightblue'])
plt.title('Perbandingan Jumlah Registrasi antara Tahun Pertama dan Tahun Kedua')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Pesanan')
plt.show()
```

### Pertanyaan 6: Bagaimana hubungan antara cuaca(weathersit) dengan jumlah sepeda yang dipinjam pada jam tertentu?
### 10122354 - Jafar Shidiq

```python
day_data = pd.read_csv("bike_day.csv")
hour_data = pd.read_csv("bike_hour.csv")
hour_data.groupby("weathersit")["cnt"].mean()

plt.figure(figsize=(10,5))
sns.barplot(x="weathersit", y="cnt", data=hour_data)
plt.title("Pengaruh Cuaca Terhadap Jumlah Sepeda Dipinjam")
plt.xlabel("Tipe Cuaca\n1: Clear, Few clouds, Partly cloudy, Partly cloudy\n2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist\n3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds\n4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog")
plt.ylabel("Rata-Rata Jumlah Sepeda")
plt.show()
```
